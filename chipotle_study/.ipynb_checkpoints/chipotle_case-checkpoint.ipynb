# Chipotle Case Study

## Define problems and/or goals
 - Find out how much, on average, people spend per order.
 - Find the most typical combination of items in the orders
 - Given a specific item, find out what is ordered most frequently with it


## Determine type of data analytics to use
 - Exploratory data analysis
 - Descriptive analytics

## Collect data to solve the problem
Data comes from the following link:
https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv?ref=datascience.fm

#### 1 ) Imports

```python
# Imports
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

#### 2 ) Read data

```python
# Read tsv file
df = pd.read_csv('chipotle_data.txt', delimiter="\t")

# Preview df
df.head()
```

## Clean/transform the data
 - Transform data in a column to a workable data type
 - Insert a calculated column to work with data

#### 1 ) Transform data in a column to a workable data type

```python
# Remove "$" signs from col
df.loc[:, "item_price"].replace("\$", "", inplace=True, regex=True)

# Change dtype of col
df = df.astype({"item_price": "float"})

# Preview dtypes
df.dtypes
```

#### 2 ) Insert a calculated column to work with data

```python
# Create an order_value col that represents total cost by quantity
df['order_value'] = df['item_price'] * df['quantity']

# Preview df
df.head()
```

## Analyze the Data

### Question 1: Find out how much, on average, people spend per order.

#### 1 ) Group entries by order number

```python
# Group together by order number
group_df = df.groupby(['order_id']).sum()

# Preview df
group_df.head()
```

#### 2 ) Search for outliers

```python
# Sort values by order value
group_df = group_df.sort_values("order_value")

# Preview df
group_df
```

#### 3 ) Remove obvious outliers

```python
# Remove outlier
group_df = group_df.drop(index=1443)

# Preview sorted df
group_df.sort_values("order_value")
```

#### 4 ) Generate descriptive statistics

```python
# Generate mean
group_df.mean(numeric_only=True)
```

```python
# Generate median
group_df.median(numeric_only=True)
```

#### Conclusion
Order value data appears right skewed but appears to average around \\$16 to \\$20.

### Question 2: Find the most typical combination of items in the orders

#### 1 ) Group unique entries together then count number of items for each group

```python
# Find most typical combinations
df.groupby("order_id")['item_name'].unique().astype(str).value_counts()
```

#### Conclusion
It appears that bowls and burritos are typically ordered along with chips and guacamole. A bowl/burrito + chips and guacamole combo seems like it could be a good business move.

### Question 3: Given a specific item, find out what is ordered most frequently with it.

#### 1 ) Perform a self join

```python
# Reset the index to create a usable index column
index_df = df.reset_index()

# Perform a "Self join" on the order ID column of the df
selfjoin_df = index_df.merge(index_df, on="order_id")

# Query the df to remove self-pairs
selfjoin_df = selfjoin_df.query("index_x > index_y")
```

#### 2 ) Isolate and sort item names and pairings

```python
# Isolate the item name columns
items_df = selfjoin_df[["item_name_x", "item_name_y"]]

# Sort the df as a np array
items_np = np.sort(items_df.to_numpy(), axis=1)

# Cast the np array back into a df
items_df = pd.DataFrame(items_np, columns=["item_name", "item_pair"])
```

#### 3 ) Group by item names and pairings, then sort by frequency of the pairing

```python
items_df.groupby([*items_df]).size().sort_values(ascending=False)
```

#### Conclusion
We now know what is often ordered with given items. This can be useful for a suggestions feature while ordering to encourage upsales.

## Present/Visualize the Data

### Order value density histogram
This histogram shows where order values are centralized such as for Q1.

```python
# Generate distribution plot of order value density
plt.figure(figsize=(10, 6))
sns.histplot(group_df["order_value"],
             kde=True, stat="density", kde_kws=dict(cut=3),
             alpha=.4, edgecolor=(1, 1, 1, .4), bins=50)
plt.xlabel("Order Value")
plt.ylabel("Frequency")
plt.title("General Distribution of the Order Value", size=24)
```

### Order combo stacked bar graph
This graph shows the most commonly ordered items along with their most popular pairings in a bar graph for comparison.

```python
items_sized = items_df.groupby([*items_df]).size()
```

```python
items_unstacked = items_sized.unstack()
items_unstacked.plot(kind="bar", stacked=True, legend = None, colormap="nipy_spectral")
plt.title("Most Common Items Combinations")
plt.xlabel("Item Name")
plt.ylabel("Order Count")
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
```

## Report/Storytelling
As stated above. From the descriptive statistics we can see that patrons order, on average, around \\$16 to \\$20 at Chipotle. The most ordered item is the Chicken Bowl and it is often comboed with the Chips and Guacamole.

### Define Other Possible Problems to Analyze
 - What items are ordered the least

### 1 ) What items are ordered the least

```python
# Group items by name, sort by amount ordered, then return only items that are ordered less than 1% as often as the most ordered item
df.groupby("item_name").sum().sort_values(by="quantity").query('quantity < (761 * 0.01)')
```

#### Conclusion
Items like those shown above are not ordered very often at all. These items should either be better advertised, altered, or removed from the menu. One way to think about it is: for every 1000 chicken bowls sold, there will be only be about 1 carnitas salad sold.
